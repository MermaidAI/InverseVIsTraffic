```julia
using JSON
```

```julia
poaDictAprPM = readall("./results/poaDictAprPM.json");
poaDictAprPM = JSON.parse(poaDictAprPM);
```

```julia
poaDictAprWeekend = readall("./results/poaDictAprWeekend.json");
poaDictAprWeekend = JSON.parse(poaDictAprWeekend);
```

```julia
poaDictAprPM
```

```julia
poaDictAprWeekend
```

```julia
poaDictApr = Dict{}()

for key in keys(poaDictAprPM)
    poaDictApr[int(key)] = poaDictAprPM[key]
end

for key in keys(poaDictAprWeekend)
    poaDictApr[int(key)] = poaDictAprWeekend[key]
end
```

```julia
poaDictApr
```

```julia
using PyPlot

day = 1:30

poa = map(day->poaDictApr[day], day)

# plot(iterNum, objFun, "s-g", label="True")

fig = figure(figsize=(8,4.8))

plot(day, poa, "o-r", linewidth=1.5)

# legend(loc="upper right",fancybox="true") 

grid("on")

xlim(0, 31)
ylim(minimum(poa)-.05, maximum(poa)+.05)

font1 = ["family"=>"serif", "color"=>"darkred", "weight"=>"normal", "size"=>12]

xlabel("April days (PM period)", fontdict=font1)
ylabel("POA", fontdict=font1)

savefig("./results/POA_Apr_ext.pdf")
```
