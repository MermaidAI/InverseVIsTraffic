```python
class TMC_Speed(object):
    def __init__(self, tmc, day, speed, travel_time):
        self.tmc = tmc
        self.day = day
        self.speed = speed
        self.travel_time = travel_time
    def ave_speed(self):
        return sum([self.speed[i] * self.travel_time[i] for i in range(len(self.speed))]) / sum(self.travel_time)
```
