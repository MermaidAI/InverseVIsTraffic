```python
link_length_journal_dict = {}
link_free_time_journal_dict = {}

number_of_links = 258

for i in range(number_of_links):
    link_length_journal_dict[str(i)] = link_length_dict_journal[str(i)].length
    link_free_time_journal_dict[str(i)] = link_length_dict_journal[str(i)].free_flow_time
```

```python
import json

# Writing JSON data
with open('../temp_files/link_length_journal_dict_JSON.json', 'w') as json_file:
    json.dump(link_length_journal_dict, json_file)
    
with open('../temp_files/link_free_time_journal_dict_JSON.json', 'w') as json_file:
    json.dump(link_free_time_journal_dict, json_file)
```

```python
with open('../temp_files/link_length_journal_dict_JSON.json', 'r') as json_file:
    link_length_journal_dict_ = json.load(json_file)
    
with open('../temp_files/link_free_time_journal_dict_JSON.json', 'r') as json_file:
    link_free_time_journal_dict_ = json.load(json_file)
```
