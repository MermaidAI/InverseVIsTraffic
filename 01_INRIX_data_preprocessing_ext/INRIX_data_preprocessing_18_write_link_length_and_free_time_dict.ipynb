```python
link_length_ext_dict = {}
link_free_time_ext_dict = {}
for i in range(64):
    link_length_ext_dict[str(i)] = link_length_dict_ext[str(i)].length
    link_free_time_ext_dict[str(i)] = link_length_dict_ext[str(i)].free_flow_time
```

```python
import json

# Writing JSON data
with open('../temp_files/link_length_ext_dict_JSON.json', 'w') as json_file:
    json.dump(link_length_ext_dict, json_file)
    
with open('../temp_files/link_free_time_ext_dict_JSON.json', 'w') as json_file:
    json.dump(link_free_time_ext_dict, json_file)
```

```python
with open('../temp_files/link_length_ext_dict_JSON.json', 'r') as json_file:
    link_length_ext_dict_ = json.load(json_file)
    
with open('../temp_files/link_free_time_ext_dict_JSON.json', 'r') as json_file:
    link_free_time_ext_dict_ = json.load(json_file)
```
