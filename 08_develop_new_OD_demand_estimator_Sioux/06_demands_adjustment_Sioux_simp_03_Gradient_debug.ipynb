```julia
include("prepare_data_Sioux_simp.jl")
```

```julia
demandsDict = Dict{Int64, Dict{(Int64,Int64), Float64}}()
# get ground trueth demands, indexed by 0
demandsDict[0] = iniDemand("../data_original/SiouxFalls_trips_simp.txt")
# get initial demands, indexed by 1
demandsDict[1] = iniDemand("../data_original/SiouxFalls_trips_simp.txt", 1)

numNodes, numLinks, numODpairs, capacity, free_flow_time, ta_data = paraNetwork("Sioux_simp")
numRoutes, odPairRoute, linkRoute, link_label_dict = furInfo()
```

```julia
include("tap_Sioux_simp.jl")
```

```julia
tapFlowVecDict = Dict{Int64, Array{Float64, 1}}()
tapFlowDicDict = Dict{Int64, Dict{(Int64,Int64),Float64}}()
```

```julia
# get observed flow vector (corresponding to ground truth demands)
tapFlowDicDict[0], tapFlowVecDict[0] = TAP(demandsDict[0])
```

```julia
# get initial flow vector (corresponding to initial demands)
tapFlowDicDict[1], tapFlowVecDict[1] = TAP(demandsDict[1])
```

```julia
require("demands_adjustment_Sioux_simp_gradi")
```

```julia
# copy observed flow vector to x_0
x_0 = tapFlowVecDict[0]

## Obtain $\left( {\frac{{\partial {c_a}\left( {{{\boldsymbol{\lambda}}^l}} \right)}}{{\partial {x_a}}}a \in \mathcal{A}} \right)$ 

fcoeffs = [1, 0, 0, 0.15]


saVecDict = Dict{Int64, Array{Float64, 1}}()
saVecDict[1] = saVect(tapFlowVecDict[1], fcoeffs, capacity, free_flow_time)
```

```julia
jacobDict = Dict{Int64, Array{Float64}}()
@time jacobDict[1] = @parallel vcat for i=1:numODpairs
    solveJacob(i, saVecDict[1], numLinks, numODpairs, numRoutes, linkRoute, odPairRoute)
end

jacobDict[1] = reshape(jacobDict[1], numODpairs, numLinks)
```
